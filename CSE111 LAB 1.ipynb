```python
#Task 1
n = input("PLEASE ENTER: ")
up = 0
low = 0
for i in n:
  if 65<=ord(i)<=90:
    up+=1
  elif 97<=ord(i)<=122:
    low+=1

if low>=up:
  print(n.lower())
else:
  print(n.upper())
```

```python
#Task 2
n = input()

num = 0
chr = 0
for i in n:
  if ord("A")<=ord(i)<=ord("Z") or ord("a")<=ord(i)<=ord("z"):
    chr+=1
  elif ord('0')<=ord(str(i))<=ord('9'):
    num+=1
if num==0:
  print("WORD")
elif chr==0:
  print("NUMBER")
else:
  print("MIXED")
```

```python
#Task 4
n = input("PLEASE ENTER:")
list1 = n.split(", ")
new=''

for i in list1[0]:
  if i in list1[1]:
    new+=i
for i in list1[1]:
  if i in list1[0]:
    new+=i
if new=='':
  print("NOTHING IN COMMON")
else:
  print(new)
```
