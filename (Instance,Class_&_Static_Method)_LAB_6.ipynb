```python
# Task 1
class Student:
  count = 0
  def __init__(self,name,dept,age,cg):
    self.name = name
    self.dept = dept
    self.age = age
    self.cg = cg
    Student.count+=1

  def showDetails(self):

    print("ID:",Student.count)
    print("Name:",self.name)
    print("Department:",self.dept)
    print("Age:",self.age)
    print("CGPA:",self.cg)

  @classmethod
  def from_String(cls,x):
    info=x.split("-")
    information= cls(info[0],info[1],int(info[2]),float(info[3]))
    return information

s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails()
```

```python
#Task 2
class Assassin:
  num_assassin = 0
  def __init__(self,name,rate):
    self.name = name
    self.rate = rate
    Assassin.num_assassin+=1
  def printDetails(self):
    print("Name:",self.name)
    print(f"Success rate:{self.rate}%")
    print("Total number of Assassin:",Assassin.num_assassin)
  @classmethod
  def failureRate(cls,name,rate):
    fRate = 100-rate
    obj1 = cls(name,fRate)
    return obj1
  @classmethod
  def failurePercentage(cls,name,rate):
    frate = 100-int(rate[:2])
    obj2 = cls(name,frate)
    return obj2

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()



```

```python
# Task 5
class Employee:
  def __init__(self,name,workingperiod):
    self.name=  name
    self.workingPeriod = workingperiod

  @classmethod
  def employeeByJoiningYear(cls,name,year):
    workingPeriod = 2023-year
    f2 = cls(name,workingPeriod)
    return f2

  @staticmethod
  def experienceCheck(workingperiod,gender):
    if workingperiod<3 and gender=="male":
      return ("He is not experienced")
    elif workingperiod<3 and gender=="female":
      return ("She is not experienced")
    elif workingperiod>3 and gender=="male":
      return ("He is experienced")
    else:
      return ("She is experienced")


employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
class Cat:
  Number_of_cats = 0
  def __init__(self,color,active):
    self.color = color
    self.active = active
    Cat.Number_of_cats+=1
  def printCat(self):
    print(f"{self.color} cat is {self.active}")
  def changeColor(self,color):
    self.color = color
  @classmethod
  def no_parameter(cls):
    color = "White"
    active = "Sitting"
    return cls(color,active)
  @classmethod
  def first_parameter(cls,color):
    active = "Sitting"
    c = cls(color,active)
    return c
  @classmethod
  def second_parameter(cls,active):
    color = "Grey"
    return cls(color,active)

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
# Task 8
import math
class Cylinder:
  radius = 5
  height = 18
  def __init__(self,rad,hgt):
    print(f"Default radius={Cylinder.radius} and height={Cylinder.height}")
    Cylinder.radius = rad
    Cylinder.height = hgt
    print(f"Updated radius={Cylinder.radius} and height={Cylinder.height}")

  @staticmethod
  def area(r,h):
    area = 2*math.pi*r*h + 2*math.pi*r**2
    print("Area:",area)

  @staticmethod
  def volume(r,h):
    vol = math.pi*r**2*h
    print("Volume:",vol)

  @classmethod
  def swap(cls,h,r):
    return cls(r,h)

  @classmethod
  def changeFormat(cls,info):
    info = info.split("-")
    rad = float(info[0])
    h = float(info[1])
    return cls(rad,h)

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)

```

```python
# TAsk 10
class SultansDine:
  num_branch = 0
  total_sell = 0
  lst = []
  def __init__(self,loc):
    self.loc = loc
    SultansDine.num_branch+=1
  def sellQuantity(self,sell):
    if sell<10:
      self.branch_sell = sell*300
    elif sell<20:
      self.branch_sell = sell*350
    else:
      self.branch_sell = sell*400
    SultansDine.total_sell+=self.branch_sell

  def branchInformation(self):
     print("Branch Name:",self.loc)
     print("Branch Sell:",self.branch_sell,"Taka")
     SultansDine.lst.append(self.loc)
     SultansDine.lst.append(self.branch_sell)

  @classmethod
  def details(cls):
    print("Total Number of branch(s):",cls.num_branch)
    print("Total Sell:",cls.total_sell,"Taka")
    for i in range(0,len(cls.lst),2):
      print(f"Branch name: {cls.lst[i]},Branch sell: {cls.lst[i+1]}")
      print(f"Branch consists of total sell's: {round((cls.lst[i+1]/cls.total_sell)*100,2)}%")

SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()


```
