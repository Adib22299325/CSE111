```python
def div_frac(x, y):
    
    if y == 0:
        return 0
    else:
        result = (x / y) - (x // y)
        if result == 0.0:
            return 0
        else:
            return result

print(div_frac(5, 2))
print(div_frac(5, 0))
print(div_frac(3, 7))
```

```python
# Task 6
def vow (name):
  name.lower()
  vowel = "aeiou"
  count= 0
  out=""
  lst = []
  for i in name:
    if i in vowel:
      count+=1
      lst.append(i)
  if len(lst)==0:
    print("NO VOWELS IN THE NAME")
  else:
    out+="Vowels: "+",".join(lst)+" .Total num of vow: "+str(count)
  return out
  
print(vow("Steve Jobs"))
```

```python
# Task 7
def palindrome_check(name):
  rev=name[::-1]
  if name==rev:
    return "Palindrome"
  else:
    return "Not a palindrome"
palindrome_check("madam")
```
