```python
# task 8
class Calculator:
  def __init__(self):
   
    self.x = int(input())
    self.op =input("Enter op:")
    self.y = int(input())
    print("Lets calculate")
    print("Value 1:",self.x)
    print("Operator:",self.op)
    print("Value 2:",self.y)
    if self.op == "+":
      self.add()
    elif self.op == "-":
      self.sub()
    elif self.op == "*":
      self.multiply()
    elif self.op == "/":
      self.division()
  def add(self):
    print(f"result: {self.x+self.y}")
  def sub(self):
    print(f"result: {self.x-self.y}")
  def multiply(self):
    print(f"result: {self.x*self.y}")
  def division(self):
    print(f"result: {self.x/self.y}")

c1 = Calculator()
```

```python
# Task 6
class box:
  def __init__(self,obj):
    print("Box 1")
    self.height=obj[0]
    self.width = obj[1]
    self.breadth = obj[2]
    print("Creating a box")
    print(f"volume of the box is {obj[0]*obj[1]*obj[2]} cubic units")


print("Box 1")
b1 = box([10,10,10])
print("=========================")
print("Height:", b1.height)
print("Width:", b1.width)
print("Breadth:", b1.breadth)
print("-------------------------")
print("Box 2")
b2 = box((30,10,10))
print("=========================")
print("Height:", b2.height)
print("Width:", b2.width)
print("Breadth:", b2.breadth)
b2.height = 300
print("Updating Box 2!")
print("Height:", b2.height)
print("Width:", b2.width)
print("Breadth:", b2.breadth)
print("-------------------------")
print("Box 3")
b3 = b2
print("Height:", b3.height)
print("Width:", b3.width)
print("Breadth:", b3.breadth)
```

```python
# TASK 9
class Patient:
 
  def __init__(self,name,age,wgt,hgt):
    self.name = name
    self.age = age
    self.wgt = wgt
    self.hgt = hgt/100
    
  def printDetails(self):
    self.bmi = self.wgt/self.hgt**2
    
    print("Name:",self.name)
    print("Age:",self.age)
    print("Weight:",self.wgt)
    print("Height:",self.hgt)
    print("BMI:",self.bmi)
p1 = Patient("A", 55, 63.0, 158.0)
p1.printDetails()
print("====================")
p2 = Patient("B", 53, 61.0, 149.0)
p2.printDetails()
```

```python
# TASK 11
class Shape:
  def __init__(self,name,b,h):
    self.name = name
    self.b = b
    self.h = h
  def area(self):
    if self.name=="Triangle":
      area = 0.5*self.b*self.h
      print("AREA:",area)
    elif self.name == "Square":
      area = self.b*self.h
      print("AREA:",area)
    elif self.name == "Rhombus":
      area = self.b*self.h*0.5
      print("AREA:",area)
    elif self.name == "Rectangle":
      area = self.b*self.h
      print("AREA:",area)
    else:
      print("SHAPE JANINA")
triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```
